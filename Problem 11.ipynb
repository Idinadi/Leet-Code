Calculate the factorial of a non-negative integer n.

Example:
Input: n = 5
Output: 120

Constraints:
- 0 <= n <= 12

```python
def factorial(input_data):
    if input_data == 0:
        return 1
    return input_data * factorial(input_data - 1)
```

```python
factorial(0)
```

```python
factorial(5)
```
