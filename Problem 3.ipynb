Given an integer x, return true if x is a palindrome, and false otherwise.

Example 1:
Input: x = 121
Output: true

Example 2:
Input: x = -121
Output: false
Explanation: From left to right, it reads -121. From right to left, it becomes 121-. Therefore it is not a palindrome.

Example 3:
Input: x = 10
Output: false

Constraints:
- -2^31 <= x <= 2^31 - 1

```python
def palindrome(x):
    if x < 0:
        return False
    original_value = x
    result = 0
    while x:
        result = result * 10 + x % 10
        x = x // 10
    return True if result == original_value else False
```

```python
palindrome(121)
```

```python
palindrome(-121)
```

```python
palindrome(10)
```
