Given a signed 32-bit integer x, return x with its digits reversed. If reversing x causes the value to go outside the signed 32-bit integer range [-2^31, 2^31 - 1], then return 0.

Example 1:
Input: x = 123
Output: 321

Example 2:
Input: x = -123
Output: -321

Example 3:
Input: x = 120
Output: 21

Constraints:
- -2^31 <= x <= 2^31 - 1

```python
def reverse(num):
    sign = -1 if num < 0 else 1
    num = abs(num)
    result = 0
    while num:
        result = result * 10 + num % 10
        num = num // 10
    result = sign * result
    return result if -2**32 <= result <= 2**32 else 0 
```

```python
reverse(123)
```

```python
reverse(-321)
```

```python
reverse(120)
```
