Every element appears twice except for one. Find that single element using O(1) extra space.

Example:
Input: nums = [2,2,1]
Output: 1

Constraints:
- 1 <= nums.length <= 3 * 10^4

```python
def singleNumber(input_data):
    result = 0
    for num in input_data:
        result ^= num
    return result
```

```python
input_data = [2, 2, 1]
singleNumber(input_data)
```

```python
input_data = [2, 3, 3, 4, 4]
singleNumber(input_data)
```
