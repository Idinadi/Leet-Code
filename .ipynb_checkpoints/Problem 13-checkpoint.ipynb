Given an array of integers, return the maximum element.

Example:
Input: nums = [3, 5, 1, 8, 2]
Output: 8

Constraints:
- 1 <= nums.length <= 1000
- -10^6 <= nums[i] <= 10^6

```python
def findMax(input_data):
    if not input_data:
        return None
    i = 0
    max = input_data[0]
    while i < len(input_data):        
        if input_data[i] > max:
            max = input_data[i]
        i += 1
    return max
```

```python
findMax([3, 5, 1, 8, 2])
```

```python
findMax([-5, -2, -9])
```

```python
findMax([])
```
