Determine if a given positive integer is prime.

Example:
Input: n = 7
Output: True

Constraints:
- 1 <= n <= 10^4

```python
def isPrime(num):
    if num <= 1:
        return False
    for i in range(2, int(num**0.5) + 1): 
        if num % i == 0:
            return False
    return True
```

```python
isPrime(1)
```

```python
isPrime(2)
```

```python
isPrime(4)
```

```python
isPrime(3)
```
