Count the number of vowels (a, e, i, o, u) in a given string (case-insensitive).

Example:
Input: s = "hello"
Output: 2

Constraints:
- 0 <= s.length <= 1000
- s consists of lowercase English letters

```python
def count_vowels(s):
    vowels = ('a', 'e', 'i', 'o', 'u')
    result = 0
    for value in s.lower():
        if value in vowels:
            result += 1
    return result
```

```python
count_vowels(s="hello")
```

```python
count_vowels(s="GAMe")
```

```python
count_vowels(s="sky")
```
