Merge two sorted arrays into one sorted array.

Example:
Input: arr1 = [1,3,5], arr2 = [2,4,6]
Output: [1,2,3,4,5,6]

Constraints:
- 0 <= arr1.length, arr2.length <= 1000

```python
def mergeSortedArrays(input_data):
    x, y = set(input_data[0]), set(input_data[1])
    return sorted(list(x) + list(y))
```

```python
input_data = [[1, 3, 5], [2, 4, 6]]
mergeSortedArrays(input_data)
```

```python
def mergeSortedArrays(input_data):
    x, y = input_data[0], input_data[1]
    i, j = 0, 0
    output = []
    while i < len(x) and j < len(y):
        print(x[i], y[j])
        if x[i] < y[j]:
            output.append(x[i])
            i += 1
        else:
            output.append(y[j])
            j += 1
        print("stage1: ", output)
    while i < len(x):
        output.append(x[i])
        i += 1
        print("stage2: ", output)
    while j < len(y):
        output.append(y[j])
        j += 1
        print("stage3: ", output)
    print(output)
```

```python
input_data = [[1, 3, 5], [2, 4, 6]]
mergeSortedArrays(input_data)
```
