Return a list for 1 to n: "FizzBuzz" if divisible by 3 and 5, "Fizz" if by 3, "Buzz" if by 5, else the number as string.

Example:
Input: n = 5
Output: ["1","2","Fizz","4","Buzz"]

Constraints:
- 1 <= n <= 10^4

```python
def fizzBuzz(input_data):
    output = []
    for num in range(input_data):
        num += 1
        if num % 3 == 0 and num % 5 == 0:
            output.append("FizzBuzz")
        elif num % 3 == 0:
            output.append("Fizz")
        elif num % 5== 0:
            output.append("Buzz")        
        else:
            output.append(str(num))
    return output
```

```python
input_data = 5
fizzBuzz(input_data)
```

```python
input_data = 15
fizzBuzz(input_data)
```
