Given an array of integers, return the maximum element.

Example:
Input: nums = [3, 5, 1, 8, 2]
Output: 8

Constraints:
- 1 <= nums.length <= 1000
- -10^6 <= nums[i] <= 10^6

```python
def solution6(input_data):
   return max(input_data)
```

```python
solution6([3, 5, 1, 8, 2])
```

```python
solution6([-5, -2, -9])
```
