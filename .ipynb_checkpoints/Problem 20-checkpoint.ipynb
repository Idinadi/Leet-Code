Find the maximum profit from one buy and one sell given daily stock prices.

Example:
Input: prices = [7,1,5,3,6,4]
Output: 5

Constraints:
- 1 <= prices.length <= 10^5

```python
def maxProfit(input_data):
    profit = 0
    min_value = float('inf')
    for value in input_data:
        if value < min_value:
            min_value = value
        else:
            profit = max(profit, value - min_value)
    return profit
```

```python
input_data = [7,1,5,3,6,4]
maxProfit(input_data)
```

```python
input_data = [2,0,4,8, 1]
maxProfit(input_data)
```
