Given a string, find the first non-repeating character and return its index. If none exists, return -1.

Example:
Input: s = "leetcode"
Output: 0

Constraints:
- 1 <= s.length <= 10^5

## Solution 1

```python
def firstUniqueChar(input_data):
    for i, s in enumerate(input_data):
        if input_data.count(s) == 1:
            return i
    return -1
```

```python
firstUniqueChar("loveleetcode")
```

```python
firstUniqueChar("z")
```

## Solution 2

```python
def firstUniqueChar(input_data):    
    j = 0
    prefix = input_data[0]
    while j <len(input_data):
        found_duplicate = False
        for i, s in enumerate(input_data[j+1:]):
            if s == prefix:
                found_duplicate = True
                break
        if found_duplicate == False:
            return j
        if j + 1 < len(input_data):
            j += 1
            prefix = input_data[j]
         
    return -1
```

```python
firstUniqueChar("loveleetcode")
```

## Solution 3

```python
def firstUniqueChar(input_data):    
    freq = {}
    for i, value in enumerate(input_data):
        freq[value] = freq.get(value, 0) + 1
    for i, value in enumerate(freq):
        if freq[value]==1:
            return i
    return -1
```
