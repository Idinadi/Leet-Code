Return an array of the number of 1s in the binary representation of each number from 0 to n.

Example:
Input: n = 5
Output: [0,1,1,2,1,2]

Constraints:
- 0 <= n <= 10^5

```python
def countBits(n):
    result = []
    for i in range(n + 1):
        result.append(bin(i).count('1'))
    return result
```

```python
countBits(5)
```

```python
countBits(7)
```

```python
countBits(15)
```
