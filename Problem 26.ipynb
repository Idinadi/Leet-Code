Compress consecutive repeated characters: "aabccc" becomes "a2bc3".

Example:
Input: s = "aabccc"
Output: "a2bc3"

Constraints:
- 1 <= s.length <= 2000

```python
def compress(input_data):
    n = input_data
    result = ""
    count = 1
    for i in range(1, len(n)):
        if n[i] == n[i-1]:
            count += 1
        else:
            result += n[i-1] + str(count)
            count = 1
    result += n[-1] + str(count)
    return result
```

```python
compress("aabccc")
```

```python
compress("aabcccccaaa")
```
