Given an array containing n distinct numbers from 0 to n, find the missing number.

Example:
Input: nums = [0,1,3]
Output: 2

Constraints:
- n == nums.length
- 1 <= n <= 10^4

```python
def missingNumber(nums):
    for i, num in enumerate(nums):
        if i != num:
            return i
    return len(nums)
```

```python
missingNumber([0,1,3])
```

```python
missingNumber([0,1])
```
