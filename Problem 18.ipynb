Given a number as an array of digits, add one to the number.

Example:
Input: digits = [1,2,3]
Output: [1,2,4]

Constraints:
- 1 <= digits.length <= 100

```python
def plusOne(input_data):
    n = len(input_data)
    for i in range(n -1, -1, -1):
        if input_data[i] < 9:
            input_data[i] += 1
            return input_data
        input_data[i] = 0
    return [1] + input_data
```

```python
input_data = [1, 2, 3]
plusOne(input_data)
```

```python
input_data = [1, 2, 9]
plusOne(input_data)
```

```python
input_data = [1, 9, 1]
plusOne(input_data)
```

```python
input_data = [1, 9, 9]
plusOne(input_data)
```

```python
input_data = [9, 9, 9]
plusOne(input_data)
```
