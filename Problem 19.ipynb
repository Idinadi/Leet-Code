Move all zeroes to the end while maintaining relative order of non-zero elements. 
Example: Input: nums = [0,1,0,3,12] Output: [1,3,12,0,0] 
Constraints: - 1 <= nums.length <= 10^4

```python
def moveZeroes(nums):
    non_zeros = [value for value in nums if value != 0]
    zeros = [value for value in nums if value == 0]
    result = non_zeros + zeros
    return result
```

```python
nums = [0, 1, 12, 3, 0]
moveZeroes(nums)
```

```python
nums = [1, 0, 13, 0, 4]
moveZeroes(nums)
```
