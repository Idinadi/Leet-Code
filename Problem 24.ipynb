Given an integer n, return true if it is a power of two.

Example:
Input: n = 16
Output: True

Constraints:
- -2^31 <= n <= 2^31 - 1

```python
def isPowerOfTwo(input_data):
    n = input_data
    if n <= 0:
        return False
    while n % 2 == 0:
        n //= 2
    return n == 1
```

```python
isPowerOfTwo(8)
```

```python
isPowerOfTwo(6)
```

```python
isPowerOfTwo(32)
```
