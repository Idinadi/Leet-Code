Calculate the nth Fibonacci number. F(0)=0, F(1)=1, F(n)=F(n-1)+F(n-2).

Example:
Input: n = 4
Output: 3

Constraints:
- 0 <= n <= 30

```python
def fib(input_data):
    if input_data <= 1:
        return input_data
    return fib(input_data - 1) + fib(input_data - 2)
```

```python
input_data = 5
fib(input_data)
```

```python
input_data = 3
fib(input_data)
```
