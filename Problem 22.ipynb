An ugly number is a positive integer whose prime factors are limited to 2, 3, and 5. Determine if a number is ugly.

Example:
Input: n = 6
Output: True

Constraints:
- -2^31 <= n <= 2^31 - 1

```python
def isUgly(input_data):
    if input_data <= 0:
        return False
    for p in [2, 3, 5]:
        while input_data % p == 0:
            input_data //= p
    return input_data == 1
```

```python
input_data = 6
isUgly(input_data)
```

```python
input_data = 30
isUgly(input_data)
```

```python
input_data = 7
isUgly(input_data)
```
