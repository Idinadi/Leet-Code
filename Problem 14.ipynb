Find the length of the longest substring without repeating characters.

Example:
Input: s = "abcabcbb"
Output: 3

Constraints:
- 0 <= s.length <= 5 * 10^4
- s consists of English letters, digits, symbols and spaces

```python
def lengthOfLongestSubstring(input_data):
    result = [[input_data[0]]]
    for value in input_data[1:]:
        if value not in result[-1]:
            result[-1].append(value)
        else:
            result.append([value])
    result = max([len("".join(item)) for item in result])
    return result
```

```python
input_data = "abcabcbb"
lengthOfLongestSubstring(input_data)
```

```python
input_data = "pweekep"
lengthOfLongestSubstring(input_data)
```

```python
input_data = "a"
lengthOfLongestSubstring(input_data)
```
