Reverse the characters of a string.

Example:
Input: s = "hello"
Output: "olleh"

Constraints:
- 1 <= s.length <= 10^5

```python
def reverseString(input_data):
    result = ""
    for m in range(len(input_data)-1, -1, -1):
        result += input_data[m]
    return result
```

```python
input_data = "hello"
reverseString(input_data)
```
