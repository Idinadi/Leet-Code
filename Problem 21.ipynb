A phrase is a palindrome if after keeping only alphanumeric characters and lowercasing, it reads the same forward and backward.

Example:
Input: s = "A man, a plan, a canal: Panama"
Output: True

Constraints:
- 1 <= s.length <= 2 * 10^5

```python
def isValidPalindrome(input_data):
    text = "".join([item.lower() for item in input_data if item.isalnum()])
    if text == text[::-1]:
        return True
    else:
        return False
```

```python
input_data = "A man, a plan, a canal: Panama"
isValidPalindrome(input_data)
```

```python
input_data = 'race a car'
isValidPalindrome(input_data)
```

```python
input_data = 'Was it a car or a cat I saw?'
isValidPalindrome(input_data)
```
