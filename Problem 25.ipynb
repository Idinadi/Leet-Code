A happy number repeatedly replaces itself by the sum of squares of its digits until it reaches 1.

Example:
Input: n = 19
Output: True

Constraints:
- 1 <= n <= 2^31 - 1

```python
def isHappy(input_data):
    n = input_data
    seen = set()
    while n != 1:
        if n in seen:
            return False
        seen.add(n)
        n = sum(int(digit) ** 2 for digit in str(n))
    return True
```

```python
isHappy(4)
```

```python
isHappy(19)
```
